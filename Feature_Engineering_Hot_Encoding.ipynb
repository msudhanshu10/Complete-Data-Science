```
df= pd.read_csv('mercedesbenz (1).csv')
```

```
df. head()
```

```
cols=['y', 'X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']
df= df[cols]
```

```
df.head()
```

```
for col in df.columns:
  print(col,':', df[col].nunique(), 'labels')
```

```
pd.get_dummies(df, drop_first=True).shape
# .get_dummies is used for hot encoding
```

188 columns will be created in this data set if we ll use Hot Encoding.
When number of features increase, it leads to curse of Dimensionallity.
So, it may impact our accuracy level.


So we will use one Hot Encoding for n most occuring labels

```
df.X1.value_counts().sort_values(ascending=False)
```

```
top_10_X1= [x for x in df.X1.value_counts().sort_values(ascending=False).head(10).index]
top_10_X1
```

```
for label in top_10_X1:
  df[label]= np.where(df['X1']==label,1,0)
```

```
df
```

```
cols=['X0', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']
top_10=[]
for col in df[cols]:
  top_10.append([x for x in df[col].value_counts().sort_values(ascending=False).head(10).index])
```

```
top_10
```

```
type(top_10[0])
```

```
i=0
for col in cols:
  for label in top_10[i]:
    df[col+'_'+label]= np.where(df[col]==label,1,0)
  i=i+1
```
