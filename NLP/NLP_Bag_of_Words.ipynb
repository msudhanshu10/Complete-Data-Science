## Step 1- Cleaning of the text.
1. Lower Sentences
2. Removing commas and other punctuation
3. Stemming/ Lammentization


```
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```
ps= PorterStemmer()
lm= WordNetLemmatizer()
```

```
sentences= nltk.sent_tokenize(paragraph)
```

```
from nltk.corpus import stopwords
```

```
## most important step for text cleaning
corpus=[]
for i in range(len(sentences)):
  review = re.sub('[^a-zA-Z]', ' ', sentences[i]) ## sub func is used to replace. ^is not symbol.
  ## Above line will replace every characted other than a-z and A-Z with space for each sentence.

  review= review.lower()
  review= review.split() ## list of words.
  review= [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
  review= ' '.join(review)
  corpus.append(review)
```

```
corpus ## sentences after removing punctuation, stop keywords, lowering and stemming
```

```
## We can use lammentization as well.

corpus=[]
for i in range(len(sentences)):
  review = re.sub('[^a-zA-Z]', ' ', sentences[i]) ## sub func is used to replace. ^is not symbol.
  ## Above line will replace every characted other than a-z and A-Z with space for each sentence.

  review= review.lower()
  review= review.split() ## list of words.
  review= [lm.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
  review= ' '.join(review)
  corpus.append(review)
```

```
## Creating Bag of Words
from sklearn.feature_extraction.text import CountVectorizer ## Does all the works of BagOfWords
cv= CountVectorizer()
X= cv.fit_transform(corpus).toarray()
```
